```python
from diffusers import StableDiffusionPipeline
import torch
```

```python
def generate_image_from_text(prompt, output_path="generated_image.png"):
    # Load the model (SD 1.4 is light; for best quality try SDXL)
    model_id = "CompVis/stable-diffusion-v1-4"
   
    pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float32)
    pipe = pipe.to("cuda" if torch.cuda.is_available() else "cpu")
   
    # Generate image
    image = pipe(prompt).images[0]
    image.save(output_path)
    print(f"Image saved at: {output_path}")
```

```python
 
# Example usage
generate_image_from_text("Indian team playing cricket")
```

```python
from modelscope.pipelines import pipeline
from modelscope.outputs import OutputKeys
from modelscope.utils.constant import Tasks

# Load the pipeline
generator = pipeline(
    task=Tasks.text_to_video_synthesis,
    model='damo/text-to-video-synthesis',
    device='cuda'  # or 'cpu' (much slower)
)

# Prompt and generate
prompt = "a dog barking"
output = generator({'text': prompt})

# Save video path
video_path = output[OutputKeys.OUTPUT_VIDEO]
print("Video saved at:", video_path)
```
